# Interregional Differences of Pediatrics Infrastructure in Korea

## Background
- Decrease in profitability of Pediatrics in Korea due to COVID-19 and increase of medical expenses
- Out of 66 Pediatric teaching hospitals in Korea, 55 has “zero” applicants in 2022
- In many regions, people are suffered from lack of Pediatrics specialists

## Goals

- Visualize inter-regional differences of Pediatrics infrastructure in Korea
- Find out poor regions that need infrastructure improvement

## Hypothesis
- 1. The medical infrastructure of Pediatrics in Korea is poor
- 2. The interregional differences of Pediatrics infrastructure exists
